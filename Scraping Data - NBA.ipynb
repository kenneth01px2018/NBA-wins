```python
import pandas as pd
import bs4
import requests
import time
```

```python
df_list = []
url = 'https://basketball.realgm.com/nba/team-stats/%i%s/Team_Totals/Regular_Season'
for i in range(2005, 2023):
    avg_url = url % (i, "/Averages")
    adv_url = url % (i, "/Advanced_Stats")
    misc_url = url % (i, "/Misc_Stats")
    avg = pd.DataFrame(pd.read_html(avg_url)[0]).drop(columns = "#")
    adv = pd.DataFrame(pd.read_html(adv_url)[0]).drop(columns = "#")
    misc = pd.DataFrame(pd.read_html(misc_url)[0]).drop(columns = "#")
    temp1 = pd.merge(avg, adv, how = 'left', on = ['Team'])
    temp2 = pd.merge(temp1, misc, how = 'left', on = ['Team'])
    temp2["Season"] = i
    df_list.append(temp2)
    time.sleep(3)
```

```python
df = pd.concat(df_list).drop(columns = ['Dbl Dbl', 'Tpl Dbl', '40 Pts', '20 Reb', '20 Ast', 
                                         '5 Stl', '5 Blk', 'High Game', 'Techs'])
df
```

```python
df = df.replace("Seattle", "Oklahoma City", regex = True)
df = df.replace("New Jersey", "Brooklyn")
```
